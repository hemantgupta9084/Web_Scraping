## Importing Necessary Libraries 

```python
import requests
import json
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

### Create Empty_list

```python
Flipkart_Phones = [] 
```

###  Scrape all Data between page 1 to 10 from Flipkart Webpage

```python
for page in range(1,11):

    # url of the webpage
    url = f"https://www.flipkart.com/search?q=laptops&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=ed76ecc2-3620-4781-ac46-d5f7c4b2e979&as-backfill=on&page={page}"
    
    # Sending HTTP Request to the website and get the response
    r = requests.get(url)
    r.status_code  #If the status_code is 200, request is succeeded. Most common status code is 200. if the request is succeded , we can get the HTML Document of the web page and start scraping.


    #Content of Webpage using content property
    html_text = r.content   

    #Converting HTML text into BeautifulSoup
    soup = BeautifulSoup(html_text,"lxml") 

    #Scrape all Data using Find_all Method
    phones = soup.find_all("div",class_="_3pLy-c row")    

    for phone in phones:
        d = {}    #Empty_Dictionary

        #Get Name of the phone
        name = "" if(phone.find("div",class_="_4rR01T")==None) else phone.find("div",class_="_4rR01T").getText()

        print(f"Collecting Data From Page: {page} And Product Name: {name}")

        #Get Price of the phone
        price = "" if(phone.find("div",class_="_30jeq3 _1_WHN1")==None) else phone.find("div",class_="_30jeq3 _1_WHN1").getText().replace("₹","").replace(",","")

        #Get Rating of the phone
        rating = "" if(phone.find("div",class_="_3LWZlK")==None) else phone.find("div",class_="_3LWZlK").getText()

        #Get Specifications of the phone
        specifications= "" if(phone.find("ul",class_="_1xgFaf")==None) else phone.find("ul",class_="_1xgFaf").getText()

        d["Name"] = name
        d["Price"] = price
        d["Rating"] = rating
        d["Specifications"] = specifications
        Flipkart_Phones.append(d)
```

### FlipKart Phones Web Scraping

```python
Flipkart_Phones
```

### Convert To Json Format

```python
f = open("Flipkart Phones Web Scraping.json","w")
json.dump(Flipkart_Phones,f,indent=4)
```

#### Import Data

```python
file = pd.DataFrame(Flipkart_Phones,columns=["Name","Price","Rating","Specifications"])
file
```

### Convert To CSV File

```python
file.to_csv('Flipkart phones Details.csv')
```
